```python
from scipy.stats import wishart

def perturb_Sigma_wishart(Sigma, T):
    df = T - 1
    scale = Sigma / df
    Sigma_perturbed = wishart.rvs(df=df, scale=scale)
    return Sigma_perturbed
```

```python
def clean_weights(w, tol=1e-10):
    w = np.asarray(w).flatten()
    w[np.abs(w) < tol] = 0.0
    return w
```

```python
def gmv_cvxpy(Sigma, solver=cp.OSQP):
    n = Sigma.shape[0]
    w = cp.Variable(n, nonneg=True)      # long-only, no shorts

    objective = cp.Minimize(cp.quad_form(w, Sigma))
    constraints = [
        cp.sum(w) == 1,                  # fully invested
    ]
    prob = cp.Problem(objective, constraints)
    prob.solve(solver=solver)
    return clean_weights(np.asarray(w.value).flatten())

    return np.asarray(w.value).flatten()


def mean_variance_cvxpy(mu, Sigma, lam=5.0, solver=cp.OSQP):
    mu    = np.asarray(mu).flatten()
    Sigma = np.asarray(Sigma)
    n = len(mu)
    w_long  = cp.Variable(n, nonneg=True)
    w_short = cp.Variable(n, nonneg=True)
    w = w_long - w_short

    objective = cp.Maximize(mu @ w - lam * cp.quad_form(w, Sigma))
    constraints = [
        cp.sum(w_long)  == 1,
        cp.sum(w_short) == 1,
    ]
    prob = cp.Problem(objective, constraints)
    prob.solve(solver=cp.OSQP)
    return clean_weights(np.asarray(w.value).flatten())

    return np.asarray(w.value).flatten()


def ridge_mv_cvxpy(mu, Sigma, lam=5.0, gamma=0.1, solver=cp.OSQP):
    # gamma increased from 0.01 to 0.1
    mu    = np.asarray(mu).flatten()
    Sigma = np.asarray(Sigma)
    n = len(mu)
    w_long  = cp.Variable(n, nonneg=True)
    w_short = cp.Variable(n, nonneg=True)
    w = w_long - w_short

    objective = cp.Maximize(
        mu @ w
        - lam   * cp.quad_form(w, Sigma)
        - gamma * cp.sum_squares(w)
    )
    constraints = [
        cp.sum(w_long)  == 1,
        cp.sum(w_short) == 1,
    ]
    prob = cp.Problem(objective, constraints)
    prob.solve(solver=cp.OSQP)
    return clean_weights(np.asarray(w.value).flatten())

    return np.asarray(w.value).flatten()


def max_sharpe_cvxpy(mu, Sigma, lam=1.0, solver=cp.SCS):
    mu    = np.asarray(mu).flatten()
    Sigma = np.asarray(Sigma)
    n = len(mu)
    w_long  = cp.Variable(n, nonneg=True)
    w_short = cp.Variable(n, nonneg=True)
    w = w_long - w_short

    objective = cp.Maximize(mu @ w - lam * cp.quad_form(w, Sigma))
    constraints = [
        cp.sum(w_long)  == 1,
        cp.sum(w_short) == 1,
    ]
    prob = cp.Problem(objective, constraints)
    prob.solve(solver=solver)
    return clean_weights(np.asarray(w.value).flatten())

    return np.asarray(w.value).flatten()


def sparse_mv_cvxpy(mu, Sigma, lam=5.0, tau=0.01, solver=cp.OSQP):
    mu    = np.asarray(mu).flatten()
    Sigma = np.asarray(Sigma)
    n = len(mu)
    w = cp.Variable(n, nonneg=True)      # long-only

    objective = cp.Maximize(
        mu @ w
        - lam * cp.quad_form(w, Sigma)
        - tau * cp.norm1(w)              # now meaningful — penalizes concentration
    )
    constraints = [
        cp.sum(w) == 1,
    ]
    prob = cp.Problem(objective, constraints)
    prob.solve(solver=solver)
    return clean_weights(np.asarray(w.value).flatten())

    return np.asarray(w.value).flatten()
```

```python
def perturb_mu_SE(mu, SE):
    eps = np.random.normal(0.0, SE, size=len(mu))   # epsilon_i ~ N(0, SE_i^2)
    return mu + eps

def fragility_L1(w0, w1, G=2.0):
    return np.sum(np.abs(w0 - w1)) / (2 * G)

# ---- settings ----
N = 1000
lam = 5.0
gamma = .1

# ---- baseline weights (compute ONCE) ----
w0_mv = mean_variance_cvxpy(mu, Sigma, lam=lam)
w0_ridge = ridge_mv_cvxpy(mu, Sigma, lam=lam, gamma=gamma)


# ---- fragility arrays (store) ----
frag_mv = np.empty(N)
frag_ridge = np.empty(N)

for k in range(N):
    mu_p = perturb_mu_SE(mu, SE)
    Sigma_p = perturb_Sigma_wishart(Sigma, T)
    w_mv_p = mean_variance_cvxpy(mu_p, Sigma_p, lam=lam)
    w_ridge_p = ridge_mv_cvxpy(mu_p, Sigma_p, lam=lam, gamma=gamma)
    frag_mv[k] = fragility_L1(w0_mv, w_mv_p,G=2.0)
    frag_ridge[k] = fragility_L1(w0_ridge, w_ridge_p)
```

```python
def equal_weight(n):
    return np.ones(n) / n

w0_ew = equal_weight(len(mu))

# Fragility is zero by definition - no inputs to perturb
# But include in box plot with array of zeros for completeness
frag_ew = np.zeros(N)
```

```python
w0_gmv = gmv_cvxpy(Sigma)

frag_gmv = np.empty(N)

for k in range(N):
    Sigma_p = perturb_Sigma_wishart(Sigma, T)
    w_p = gmv_cvxpy(Sigma_p)
    frag_gmv[k] = fragility_L1(w0_gmv, w_p)
```

```python
def max_sharpe_like_cvxpy(mu, Sigma, lam=1.0, solver=cp.SCS):
    """
    Dollar-neutral max Sharpe-like portfolio
    Maximizes: mu'w - lam * w'Σw
    Subject to: sum(w) = 0, ||w||_1 <= 2
    """
    mu = np.asarray(mu).flatten()
    Sigma = np.asarray(Sigma)

    n = len(mu)
    w = cp.Variable(n)

    objective = cp.Maximize(mu @ w - lam * cp.quad_form(w, Sigma))
    constraints = [
        cp.sum(w) == 0,           # dollar neutral
        cp.norm1(w) <= 2.0,       # gross exposure <= 200%
    ]

    prob = cp.Problem(objective, constraints)
    prob.solve(solver=solver)
    return clean_weights(np.asarray(w.value).flatten())


    return np.asarray(w.value).flatten()
```

```python
w0_sharpe = max_sharpe_like_cvxpy(mu, Sigma, lam=1.0)

frag_sharpe = np.empty(N)

for k in range(N):
    mu_p = perturb_mu_SE(mu, SE)
    Sigma_p = perturb_Sigma_wishart(Sigma, T)
    w_p = max_sharpe_like_cvxpy(mu_p, Sigma_p, lam=1.0)
    frag_sharpe[k] = fragility_L1(w0_sharpe, w_p,G=2.0)
```

```python
w0_sparse = sparse_mv_cvxpy(mu, Sigma, lam=5.0, tau=0.1)

frag_sparse = np.empty(N)

for k in range(N):
    mu_p = perturb_mu_SE(mu, SE)
    Sigma_p = perturb_Sigma_wishart(Sigma, T)
    w_p = sparse_mv_cvxpy(mu_p, Sigma_p, lam=5.0, tau=0.1)
    frag_sparse[k] = fragility_L1(w0_sparse, w_p)
```

```python
from sklearn.covariance import LedoitWolf

# Compute shrunk covariance
lw = LedoitWolf().fit(returns_df.values)
Sigma_shrunk = lw.covariance_

# Recompute SE using shrunk covariance diagonal
sigma_assets_shrunk = np.sqrt(np.diag(Sigma_shrunk))
SE_shrunk = sigma_assets_shrunk / np.sqrt(T)

# Recompute baseline weights for each model using shrunk covariance
w0_mv_shrunk = mean_variance_cvxpy(mu, Sigma_shrunk, lam=lam)
w0_ridge_shrunk = ridge_mv_cvxpy(mu, Sigma_shrunk, lam=lam, gamma=gamma)
w0_gmv_shrunk = gmv_cvxpy(Sigma_shrunk)
w0_sharpe_shrunk = max_sharpe_like_cvxpy(mu, Sigma_shrunk, lam=1.0)
w0_sparse_shrunk = sparse_mv_cvxpy(mu, Sigma_shrunk, lam=5.0, tau=0.1)

# Run perturbation loops for each model using shrunk covariance
# Same structure as Change 2 but with Sigma_shrunk as baseline
# and perturb_Sigma_wishart(Sigma_shrunk, T) for covariance perturbation

# Example for MVO with shrinkage:
frag_mv_shrunk = np.empty(N)
frag_ridge_shrunk = np.empty(N)

for k in range(N):
    mu_p = perturb_mu_SE(mu, SE_shrunk)
    Sigma_p = perturb_Sigma_wishart(Sigma_shrunk, T)
    w_mv_p = mean_variance_cvxpy(mu_p, Sigma_p, lam=lam)
    w_ridge_p = ridge_mv_cvxpy(mu_p, Sigma_p, lam=lam, gamma=gamma)
    frag_mv_shrunk[k] = fragility_L1(w0_mv_shrunk, w_mv_p)
    frag_ridge_shrunk[k] = fragility_L1(w0_ridge_shrunk, w_ridge_p)
```

```python
import matplotlib.pyplot as plt

# GMV with shrinkage
frag_gmv_shrunk = np.empty(N)
for k in range(N):
    Sigma_p = perturb_Sigma_wishart(Sigma_shrunk, T)
    w_gmv_p = gmv_cvxpy(Sigma_p)
    frag_gmv_shrunk[k] = fragility_L1(w0_gmv_shrunk, w_gmv_p)

# Max Sharpe with shrinkage
frag_sharpe_shrunk = np.empty(N)
for k in range(N):
    mu_p = perturb_mu_SE(mu, SE_shrunk)
    Sigma_p = perturb_Sigma_wishart(Sigma_shrunk, T)
    w_sharpe_p = max_sharpe_like_cvxpy(mu_p, Sigma_p, lam=1.0)
    frag_sharpe_shrunk[k] = fragility_L1(w0_sharpe_shrunk, w_sharpe_p)

# Sparse MV with shrinkage
frag_sparse_shrunk = np.empty(N)
for k in range(N):
    mu_p = perturb_mu_SE(mu, SE_shrunk)
    Sigma_p = perturb_Sigma_wishart(Sigma_shrunk, T)
    w_sparse_p = sparse_mv_cvxpy(mu_p, Sigma_p, lam=5.0, tau=0.1)
    frag_sparse_shrunk[k] = fragility_L1(w0_sparse_shrunk, w_sparse_p)


# Create side-by-side box plots comparing raw vs shrunk covariance
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Portfolio Fragility: Raw vs Shrunk Covariance', fontsize=16, fontweight='bold')

models = [
    ('Mean-Variance', frag_mv, frag_mv_shrunk),
    ('Ridge MV', frag_ridge, frag_ridge_shrunk),
    ('GMV', frag_gmv, frag_gmv_shrunk),
    ('Max Sharpe', frag_sharpe, frag_sharpe_shrunk),
    ('Sparse MV', frag_sparse, frag_sparse_shrunk),
]

for idx, (name, frag_raw, frag_shrunk) in enumerate(models):
    row = idx // 3
    col = idx % 3
    ax = axes[row, col]

    # Create box plot
    bp = ax.boxplot([frag_raw, frag_shrunk],
                     labels=['Raw', 'Shrunk'],
                     patch_artist=True,
                     widths=0.6)

    # Color the boxes
    colors = ['lightcoral', 'lightblue']
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)

    ax.set_title(name, fontweight='bold')
    ax.set_ylabel('Fragility (L1 norm)')
    ax.grid(True, alpha=0.3)

    # Add mean values as text
    mean_raw = np.mean(frag_raw)
    mean_shrunk = np.mean(frag_shrunk)
    ax.text(1, mean_raw, f'{mean_raw:.3f}', ha='center', va='bottom', fontsize=8)
    ax.text(2, mean_shrunk, f'{mean_shrunk:.3f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()

# Optional: Print summary statistics
print("\n=== Fragility Summary: Raw vs Shrunk Covariance ===")
print(f"{'Model':<15} {'Raw Mean':<12} {'Shrunk Mean':<12} {'Reduction %':<12}")
print("-" * 55)
for name, frag_raw, frag_shrunk in models:
    mean_raw = np.mean(frag_raw)
    mean_shrunk = np.mean(frag_shrunk)
    reduction = (mean_raw - mean_shrunk) / mean_raw * 100
    print(f"{name:<15} {mean_raw:<12.4f} {mean_shrunk:<12.4f} {reduction:<12.2f}")
```
