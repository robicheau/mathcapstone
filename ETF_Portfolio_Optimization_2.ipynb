```python
def mean_variance_cvxpy(mu, Sigma, lam=5.0, solver=cp.ECOS):
    mu = np.asarray(mu).flatten()
    Sigma = np.asarray(Sigma)

    n = len(mu)
    w = cp.Variable(n)

    objective = cp.Maximize(mu @ w - lam * cp.quad_form(w, Sigma))
    constraints = [
    cp.sum(w) == 0,           # dollar neutral
    cp.norm1(w) <= 2.0,       # gross exposure <= 200%
]


    prob = cp.Problem(objective, constraints)
    prob.solve(solver=cp.OSQP)

    return np.asarray(w.value).flatten()

def ridge_mv_cvxpy(mu, Sigma, lam=5.0, gamma=1.0, solver=cp.ECOS):
    mu = np.asarray(mu).flatten()
    Sigma = np.asarray(Sigma)

    n = len(mu)
    w = cp.Variable(n)

    objective = cp.Maximize(
        mu @ w
        - lam * cp.quad_form(w, Sigma)
        - gamma * cp.sum_squares(w)
    )
    constraints = [
        cp.sum(w) == 0,
        cp.norm1(w) <= 2.0,
    ]


    prob = cp.Problem(objective, constraints)
    prob.solve(solver=cp.OSQP)

    return np.asarray(w.value).flatten()
```

```python
def perturb_mu_SE(mu, SE):
    eps = np.random.normal(0.0, SE, size=len(mu))   # epsilon_i ~ N(0, SE_i^2)
    return mu + eps

def fragility_L1(w0, w1):
    return 0.5 * np.sum(np.abs(w0 - w1))

# ---- settings ----
N = 1000
lam = 5.0
gamma = .01

# ---- baseline weights (compute ONCE) ----
w0_mv = mean_variance_cvxpy(mu, Sigma, lam=lam)
w0_ridge = ridge_mv_cvxpy(mu, Sigma, lam=lam, gamma=gamma)


# ---- fragility arrays (store) ----
frag_mv = np.empty(N)
frag_ridge = np.empty(N)

for k in range(N):
    mu_p = perturb_mu_SE(mu, SE)
    Sigma_p = perturb_Sigma_wishart(Sigma, T)
    w_mv_p = mean_variance_cvxpy(mu_p, Sigma_p, lam=lam)
    w_ridge_p = ridge_mv_cvxpy(mu_p, Sigma_p, lam=lam, gamma=gamma)
    frag_mv[k] = fragility_L1(w0_mv, w_mv_p)
    frag_ridge[k] = fragility_L1(w0_ridge, w_ridge_p)


print("MV:    mean", frag_mv.mean(), "  var", frag_mv.var(), "  p95", np.percentile(frag_mv, 95))
print("Ridge: mean", frag_ridge.mean(), "  var", frag_ridge.var(), "  p95", np.percentile(frag_ridge, 95))
```

```python
def gmv_cvxpy(Sigma, solver=cp.OSQP):
    n = Sigma.shape[0]
    w = cp.Variable(n)

    objective = cp.Minimize(cp.quad_form(w, Sigma))
    constraints = [
    cp.sum(w) == 0,           # dollar neutral
    cp.norm1(w) <= 2.0,       # gross exposure <= 200%
]

    prob = cp.Problem(objective, constraints)
    prob.solve(solver=solver)

    return np.asarray(w.value).flatten()
```

```python
from scipy.stats import wishart

def perturb_Sigma_wishart(Sigma, T):
    df = T - 1
    scale = Sigma / df
    Sigma_perturbed = wishart.rvs(df=df, scale=scale)
    return Sigma_perturbed
```

```python
from scipy.optimize import minimize

def risk_parity(Sigma):
    n = Sigma.shape[0]

    def risk_contributions(w):
        port_var = w @ Sigma @ w
        marginal_contrib = Sigma @ w
        risk_contrib = w * marginal_contrib / np.sqrt(port_var)
        return risk_contrib

    def objective(w):
        rc = risk_contributions(w)
        target = np.sqrt(w @ Sigma @ w) / n
        return np.sum((rc - target) ** 2)

    w0 = np.ones(n) / n
    bounds = [(0.0, 1.0)] * n
    constraints = {'type': 'eq', 'fun': lambda w: np.sum(w) - 1.0}

    result = minimize(objective, w0, method='SLSQP',
                      bounds=bounds, constraints=constraints)
    return result.x

w0_rp = risk_parity(Sigma)

frag_rp = np.empty(N)
for k in range(N):
    Sigma_p = perturb_Sigma_wishart(Sigma, T)
    w_p = risk_parity(Sigma_p)
    frag_rp[k] = fragility_L1(w0_rp, w_p)
```

```python
def max_sharpe_cvxpy(mu, Sigma, rf=0.0):
    n = len(mu)
    # Reformulation: maximize (mu-rf)'y subject to
    # y'Σy <= 1, sum(y) >= 0, y >= 0
    # then w = y / sum(y)
    y = cp.Variable(n)
    objective = cp.Maximize((mu - rf) @ y)
    constraints = [
        cp.quad_form(y, Sigma) <= 1,
        cp.sum(y) >= 0,
        y >= 0
    ]
    prob = cp.Problem(objective, constraints)
    prob.solve(solver=cp.SCS)
    weights = y.value / np.sum(y.value)
    return np.asarray(weights).flatten()
```

```python
w0_sharpe = max_sharpe_like_cvxpy(mu, Sigma, lam=1.0)

frag_sharpe = np.empty(N)

for k in range(N):
    mu_p = perturb_mu_SE(mu, SE)
    Sigma_p = perturb_Sigma_wishart(Sigma, T)
    w_p = max_sharpe_like_cvxpy(mu_p, Sigma_p, lam=1.0)
    frag_sharpe[k] = fragility_L1(w0_sharpe, w_p)
```

```python
def sparse_mv_cvxpy(mu, Sigma, lam=5.0, tau=0.1, solver=cp.OSQP):
    n = len(mu)
    w = cp.Variable(n)

    objective = cp.Maximize(
        mu @ w
        - lam * cp.quad_form(w, Sigma)
        - tau * cp.norm1(w)
    )

    constraints = [
    cp.sum(w) == 0,           # dollar neutral
    cp.norm1(w) <= 2.0,       # gross exposure <= 200%
]


    prob = cp.Problem(objective, constraints)
    prob.solve(solver=solver)

    return np.asarray(w.value).flatten()
```

```python
w0_sparse = sparse_mv_cvxpy(mu, Sigma, lam=5.0, tau=0.1)

frag_sparse = np.empty(N)

for k in range(N):
    mu_p = perturb_mu_SE(mu, SE)
    Sigma_p = perturb_Sigma_wishart(Sigma, T)
    w_p = sparse_mv_cvxpy(mu_p, Sigma_p, lam=5.0, tau=0.1)
    frag_sparse[k] = fragility_L1(w0_sparse, w_p)
```
