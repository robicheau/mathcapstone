```python
files = {
    "VTI":  "VTI_raw.csv",
    "IXUS": "IXUS_raw.csv",
    "AGG":  "AGG_raw.csv",
    "VNQ":  "VNQ_raw.csv",
    "DBC":  "DBC_raw.csv",
}

def load_monthly_returns(csv_path):
    df = pd.read_csv(csv_path)
    df = df.rename(columns={"Close/Last": "Close"})
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values("Date").set_index("Date")

    monthly_price = df["Close"].resample("M").last()
    monthly_ret = monthly_price.pct_change().dropna()
    return monthly_ret

monthly_returns = {ticker: load_monthly_returns(path) for ticker, path in files.items()}

returns_df = pd.DataFrame(monthly_returns).dropna()
print("returns_df shape:", returns_df.shape)
returns_df.head()
```

```python
mu = returns_df.mean().values               # (n,)
Sigma = returns_df.cov().values             # (n,n)

sigma_assets = returns_df.std().values      # sample std dev per asset
T = len(returns_df)                         # sample size (# months)
SE = sigma_assets / np.sqrt(T)              # standard error of mean per asset
```

```python
def mean_variance_cvxpy(mu, Sigma, lam=5.0, solver=cp.ECOS):
    mu = np.asarray(mu).flatten()
    Sigma = np.asarray(Sigma)

    n = len(mu)
    w = cp.Variable(n)

    objective = cp.Maximize(mu @ w - lam * cp.quad_form(w, Sigma))
    constraints = [cp.sum(w) == 1, w >= 0]

    prob = cp.Problem(objective, constraints)
    prob.solve(solver=cp.OSQP)

    return np.asarray(w.value).flatten()

def ridge_mv_cvxpy(mu, Sigma, lam=5.0, gamma=1.0, solver=cp.ECOS):
    mu = np.asarray(mu).flatten()
    Sigma = np.asarray(Sigma)

    n = len(mu)
    w = cp.Variable(n)

    objective = cp.Maximize(
        mu @ w
        - lam * cp.quad_form(w, Sigma)
        - gamma * cp.sum_squares(w)
    )
    constraints = [cp.sum(w) == 1, w >= 0]

    prob = cp.Problem(objective, constraints)
    prob.solve(solver=cp.OSQP)

    return np.asarray(w.value).flatten()
```

```python
def perturb_mu_SE(mu, SE):
    eps = np.random.normal(0.0, SE, size=len(mu))   # epsilon_i ~ N(0, SE_i^2)
    return mu + eps

def fragility_L1(w0, w1):
    return 0.5 * np.sum(np.abs(w0 - w1))

# ---- settings ----
N = 1000
lam = 5.0
gamma = .01

# ---- baseline weights (compute ONCE) ----
w0_mv = mean_variance_cvxpy(mu, Sigma, lam=lam)
w0_ridge = ridge_mv_cvxpy(mu, Sigma, lam=lam, gamma=gamma)


# ---- fragility arrays (store) ----
frag_mv = np.empty(N)
frag_ridge = np.empty(N)

for k in range(N):
    mu_p = perturb_mu_SE(mu, SE)

    w_mv_p = mean_variance_cvxpy(mu_p, Sigma, lam=lam)
    w_ridge_p = ridge_mv_cvxpy(mu_p, Sigma, lam=lam, gamma=gamma)

    frag_mv[k] = fragility_L1(w0_mv, w_mv_p)
    frag_ridge[k] = fragility_L1(w0_ridge, w_ridge_p)

print("MV:    mean", frag_mv.mean(), "  var", frag_mv.var(), "  p95", np.percentile(frag_mv, 95))
print("Ridge: mean", frag_ridge.mean(), "  var", frag_ridge.var(), "  p95", np.percentile(frag_ridge, 95))
```

```python
import matplotlib.pyplot as plt
data = [frag_mv, frag_ridge]
labels = ["Mean-Variance", f"Ridge MV (gamma={gamma})"]

plt.figure(figsize=(8,6))
plt.boxplot(data, tick_labels=labels)
plt.ylabel("Fragility (0.5 * L1 distance)")
plt.title("Fragility Distributions: MV vs Ridge MV")
plt.grid(alpha=0.3)
plt.xticks(rotation=10)
plt.show()
```

```python
def gmv_cvxpy(Sigma, solver=cp.OSQP):
    n = Sigma.shape[0]
    w = cp.Variable(n)

    objective = cp.Minimize(cp.quad_form(w, Sigma))
    constraints = [
        cp.sum(w) == 1,
        w >= 0
    ]

    prob = cp.Problem(objective, constraints)
    prob.solve(solver=solver)

    return np.asarray(w.value).flatten()
```

```python
def perturb_Sigma(Sigma, pct=0.10):
    n = Sigma.shape[0]

    noise = np.random.normal(0, pct, size=(n, n))
    noise = 0.5 * (noise + noise.T)   # make symmetric

    Sigma_p = Sigma * (1 + noise)

    # ensure positive‑semidefinite
    vals, vecs = np.linalg.eigh(Sigma_p)
    vals = np.maximum(vals, 1e-6)
    Sigma_p = vecs @ np.diag(vals) @ vecs.T

    return Sigma_p
```

```python
w0_gmv = gmv_cvxpy(Sigma)

frag_gmv = np.empty(N)

for k in range(N):
    Sigma_p = perturb_Sigma(Sigma)   # your covariance perturbation
    w_p = gmv_cvxpy(Sigma_p)
    frag_gmv[k] = fragility_L1(w0_gmv, w_p)
```

```python
def max_sharpe_like_cvxpy(mu, Sigma, lam=1.0, solver=cp.OSQP):
    n = len(mu)
    w = cp.Variable(n)

    objective = cp.Maximize(mu @ w - lam * cp.quad_form(w, Sigma))

    constraints = [
        cp.sum(w) == 1,
        w >= 0
    ]

    prob = cp.Problem(objective, constraints)
    prob.solve(solver=solver)

    return np.asarray(w.value).flatten()
```

```python
w0_sharpe = max_sharpe_like_cvxpy(mu, Sigma, lam=1.0)

frag_sharpe = np.empty(N)

for k in range(N):
    mu_p = perturb_mu_SE(mu, SE)
    w_p = max_sharpe_like_cvxpy(mu_p, Sigma, lam=1.0)
    frag_sharpe[k] = fragility_L1(w0_sharpe, w_p)
```

```python
def sparse_mv_cvxpy(mu, Sigma, lam=5.0, tau=0.1, solver=cp.OSQP):
    n = len(mu)
    w = cp.Variable(n)

    objective = cp.Maximize(
        mu @ w
        - lam * cp.quad_form(w, Sigma)
        - tau * cp.norm1(w)
    )

    constraints = [
        cp.sum(w) == 1,
        w >= 0
    ]

    prob = cp.Problem(objective, constraints)
    prob.solve(solver=solver)

    return np.asarray(w.value).flatten()
```

```python
w0_sparse = sparse_mv_cvxpy(mu, Sigma, lam=5.0, tau=0.1)

frag_sparse = np.empty(N)

for k in range(N):
    mu_p = perturb_mu_SE(mu, SE)
    w_p = sparse_mv_cvxpy(mu_p, Sigma, lam=5.0, tau=0.1)
    frag_sparse[k] = fragility_L1(w0_sparse, w_p)
```

```python
import matplotlib.pyplot as plt

data = [
    frag_ridge,
    frag_gmv,
    frag_sparse,
    frag_sharpe,
    frag_mv
]

labels = [
    "Ridge",
    "GMV",
    "Sparse MV",
    "Max Sharpe",
    "Mean‑Variance"
]

plt.figure(figsize=(10,6))

plt.boxplot(data, tick_labels=labels, showfliers=True)

plt.ylabel("Fragility (0.5 * L1 distance)")
plt.title("Fragility Distribution Comparison Across Models")
plt.grid(alpha=0.3)

plt.xticks(rotation=20)

plt.show()
```
